# Immutable argument (integer)

Here's a Python program that demonstrates the behavior of **mutable** and **immutable** arguments in functions:

```python
def modify_values(immutable_var, mutable_var):
    # Trying to modify an immutable variable (int)
    immutable_var += 10
    print("Inside function (immutable):", immutable_var)

    # Modifying a mutable variable (list)
    mutable_var.append(100)
    print("Inside function (mutable):", mutable_var)


# Immutable argument (integer)
num = 50

# Mutable argument (list)
my_list = [1, 2, 3]

# Calling the function
modify_values(num, my_list)

# Printing values after function call
print("Outside function (immutable):", num)  # Unchanged
print("Outside function (mutable):", my_list)  # Changed
```

### Explanation:
1. **Immutable objects** (like integers, strings, and tuples) **do not change** when passed to a function. A new copy is created instead.
2. **Mutable objects** (like lists and dictionaries) **can be modified** inside a function, and the changes persist outside the function.

# Initialize variables
reversed_string = ""
index = len(string) - 1  # Start from the last character

    reversed_string += string[index]
    index -= 1  # Move to the previous character

# Print the reversed string
print("Reversed String:", reversed_string)
Here's a Python program to **reverse a string** using a **while loop**:

```python
# Get user input
string = input("Enter a string: ")

# Initialize variables
reversed_string = ""
index = len(string) - 1  # Start from the last character

# Reverse the string using a while loop
while index >= 0:
    reversed_string += string[index]
    index -= 1  # Move to the previous character

# Print the reversed string
print("Reversed String:", reversed_string)
```

### Explanation:
- We start from the **last character** of the string.
- We **append** each character to `reversed_string` in reverse order.
- The **loop continues** until we reach the **first character**.

15. Write a Python program to calculate the factorial of a number provided by the user using a while loop.
-   # Get user input
num = int(input("Enter a number: "))

# Initialize factorial and counter
factorial = 1
i = num

# Calculate factorial using a while loop
while i > 0:
    factorial *= i
    i -= 1  # Decrease counter

# Print the result
print(f"Factorial of {num} is: {factorial}")
Here's a Python program to calculate the **factorial** of a number using a **while loop**:

```python
# Get user input
num = int(input("Enter a number: "))

# Initialize factorial and counter
factorial = 1
i = num

# Calculate factorial using a while loop
while i > 0:
    factorial *= i
    i -= 1  # Decrease counter

# Print the result
print(f"Factorial of {num} is: {factorial}")
```

### Explanation:
1. **Initialize `factorial = 1`** and set `i = num`.
2. Multiply `factorial` by `i`, then **decrease `i`** in each iteration.
3. **Loop stops** when `i` becomes `0`.
